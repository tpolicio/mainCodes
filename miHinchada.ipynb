```python
import json
import string
import tweepy

# claves de acceso
keysAPI = json.load(open('keysAPI.txt','r'))
consumer_key = keysAPI['consumer_key']
consumer_secret = keysAPI['consumer_secret']
access_token = keysAPI['access_token']
access_token_secret = keysAPI['access_token_secret']

# preparando:
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api=tweepy.API(auth, wait_on_rate_limit=True,
               wait_on_rate_limit_notify=True,
               parser=tweepy.parsers.JSONParser())
```

```python
# id only:
folMAGA=[]
for fo in tweepy.Cursor(api.followers_ids, screen_name="@magallanesjm").pages():
    folMAGA.extend(fo['ids'])
```

```python
len(folMAGA)
```

```python
from datetime import datetime
from datetime import date
from dateutil import relativedelta

today = date.today()
dateT=lambda t:datetime.strptime(t['created_at'],'%a %b %d %H:%M:%S +0000 %Y').date()

folMAGAall=[]
bad_folMAGA=[]
for follower in folMAGA:
    try:
        user = api.get_user(follower)
        idUser=user['id_str']
        alias=user['screen_name']
        fans=user['followers_count']
        fanOf=user['friends_count']
        favor=user['favourites_count']
        statu=user['statuses_count']
        duration=relativedelta.relativedelta(today, dateT(user))
        folMAGAall.append([user['name'],idUser,alias,fans,fanOf,favor,statu,duration.years,duration.months])
    except:
        bad_folMAGA.append(follower)
```

```python
import pandas as pd

folMAGA_DF=pd.DataFrame(folMAGAall,columns=["nombre","id_user","alias","seguido_Por","siguiendo_A","fav","statu","a침os_enTwitter","meses_enTwitter"])
folMAGA_DF
```
